```python
#import libs
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#device using cuda or cpu -> prioritize cuda
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
#import data
class CustomMNISTDataset(Dataset):
  def __init__(self, csv_file, transform=None, isTest = None):
    self.data = pd.read_csv(csv_file)
    self.transform = transform
    self.is_test = isTest

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    item = self.data.iloc[idx]

    if self.is_test:
      image = item.values.reshape(28, 28).astype(np.uint8)
      label = None
    else:
      image = item[1:].values.reshape(28, 28).astype(np.uint8)
      label = item.iloc[0]

    image = transforms.ToPILImage()(image)
    #then check none
    if self.transform is not None:
      image = self.transform(image)
    #test the image before return
    if self.is_test:
      return image
    else:
      if isinstance(label, (int, float)):
          label = torch.tensor(label)
      return image, label
```

```python
#transform the data
transform = transforms.Compose([
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
#train data
train_data = CustomMNISTDataset(csv_file='train.csv', transform=transform, isTest = False)
test_data = CustomMNISTDataset(csv_file='test.csv', transform=transform, isTest = True)
```
